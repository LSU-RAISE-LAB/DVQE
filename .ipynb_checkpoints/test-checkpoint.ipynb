```python
import numpy as np
import itertools
import time
import pandas as pd

from qiskit_aer import AerSimulator
from raiselab.core import dvqe

# ============================================================
# USER SETTINGS
# ============================================================

n1 = 3
n2 = 15
n3 = 10   # number of DVQE runs per n

mode = "distributed"

base_seed = 100
density = 0.35

verbose = False

save_csv = True
raw_results_csv = "dvqe_raw_results.csv"
summary_csv = "dvqe_summary_results.csv"


# ============================================================
# FAST BACKEND
# ============================================================
# Keeps the distributed backbone unchanged.
# Only changes the simulator backend.

fast_backend = AerSimulator(
    method="statevector",
    max_parallel_threads=0
)

# Alternative backend to test later:
#
# fast_backend = AerSimulator(
#     method="matrix_product_state",
#     max_parallel_threads=0
# )


# ============================================================
# ADAPTIVE HYPERPARAMETERS
# ============================================================

def get_hyperparameters(n):
    """
    Adaptive settings:
    - small n: cheaper settings
    - medium n: balanced settings
    - large n: stronger training settings
    """

    if n <= 8:
        return {
            "depth": 1,
            "init_type": 2,              # 1=random, 2=Black Hole, 3=GWO, 4=ABC
            "lr": 0.08,
            "max_iters": 30,
            "rel_tol": 1e-4,
            "num_shots": 128,
            "final_shots": 1000,
            "warm_start_population": 4,
            "warm_start_iters": 5,
            "warm_start_shots": 64,
            "energy_mode": "cvar",
            "cvar_alpha": 0.2,
        }

    elif n <= 12:
        return {
            "depth": 2,
            "init_type": 2,
            "lr": 0.08,
            "max_iters": 70,
            "rel_tol": 1e-5,
            "num_shots": 256,
            "final_shots": 3000,
            "warm_start_population": 5,
            "warm_start_iters": 10,
            "warm_start_shots": 128,
            "energy_mode": "cvar",
            "cvar_alpha": 0.2,
        }

    else:
        return {
            "depth": 2,
            "init_type": 2,
            "lr": 0.05,
            "max_iters": 150,
            "rel_tol": 1e-5,
            "num_shots": 512,
            "final_shots": 8000,
            "warm_start_population": 10,
            "warm_start_iters": 25,
            "warm_start_shots": 256,
            "energy_mode": "cvar",
            "cvar_alpha": 0.1,
        }


# ============================================================
# QPU CONFIG GENERATOR
# ============================================================

def make_qpu_config(n, max_qubits_per_qpu=5):
    """
    Automatically creates a distributed QPU config for each n.
    Example:
    n=5  -> [5]
    n=8  -> [5, 3]
    n=12 -> [5, 5, 2]
    """
    config = []
    remaining = n

    while remaining > 0:
        q = min(max_qubits_per_qpu, remaining)
        config.append(q)
        remaining -= q

    return config


# ============================================================
# BRUTE-FORCE EXACT QUBO SOLVER
# ============================================================

def brute_force_qubo(Q, q_linear):
    n = len(q_linear)
    best_cost = float("inf")
    best_solutions = []

    for bits in itertools.product([0, 1], repeat=n):
        z = np.array(bits, dtype=int)
        cost = float(z.T @ Q @ z + q_linear @ z)

        if cost < best_cost:
            best_cost = cost
            best_solutions = [z.copy()]
        elif abs(cost - best_cost) < 1e-9:
            best_solutions.append(z.copy())

    return best_solutions[0], best_cost, best_solutions


# ============================================================
# RANDOM QUBO GENERATOR WITH UNIQUE EXACT SOLUTION
# ============================================================

def create_random_qubo_unique_solution(
    n,
    seed=1,
    density=0.35,
    max_tries=500,
    value_low=-5,
    value_high=6
):
    for attempt in range(max_tries):
        rng = np.random.default_rng(seed + attempt)

        Q = rng.integers(value_low, value_high, size=(n, n)).astype(float)

        Q = np.triu(Q)
        Q = Q + Q.T - np.diag(np.diag(Q))

        mask = (rng.random((n, n)) < density).astype(float)
        mask = np.triu(mask)
        mask = mask + mask.T - np.diag(np.diag(mask))

        Q = Q * mask

        q_linear = rng.integers(value_low, value_high, size=n).astype(float)

        z_exact, cost_exact, all_best = brute_force_qubo(Q, q_linear)

        if len(all_best) == 1:
            return Q, q_linear, z_exact, cost_exact, attempt + 1

    raise RuntimeError(f"Could not find a unique-solution QUBO for n={n}.")


# ============================================================
# OPTIONAL LOCAL BIT-FLIP POLISHING
# ============================================================

def local_bitflip_search(z, Q, q_linear, max_passes=20):
    """
    Greedy local improvement for QUBO.

    Starting from the DVQE bitstring, flip one bit at a time if it improves
    the QUBO cost. This does not change the distributed quantum backbone.
    It only polishes the final sampled solution.
    """
    z = np.array(z, dtype=int).copy()
    best_cost = float(z.T @ Q @ z + q_linear @ z)

    n = len(z)

    for _ in range(max_passes):
        improved = False

        for i in range(n):
            z_trial = z.copy()
            z_trial[i] = 1 - z_trial[i]

            cost_trial = float(z_trial.T @ Q @ z_trial + q_linear @ z_trial)

            if cost_trial < best_cost:
                z = z_trial
                best_cost = cost_trial
                improved = True

        if not improved:
            break

    return z, best_cost


# ============================================================
# MAIN RUNNER
# ============================================================

raw_results = []
summary_results = []

overall_start = time.time()

for n in range(n1, n2 + 1):

    hp = get_hyperparameters(n)

    depth = hp["depth"]
    init_type = hp["init_type"]
    lr = hp["lr"]
    max_iters = hp["max_iters"]
    rel_tol = hp["rel_tol"]
    num_shots = hp["num_shots"]
    final_shots = hp["final_shots"]
    warm_start_population = hp["warm_start_population"]
    warm_start_iters = hp["warm_start_iters"]
    warm_start_shots = hp["warm_start_shots"]
    energy_mode = hp["energy_mode"]
    cvar_alpha = hp["cvar_alpha"]

    print("\n" + "=" * 80)
    print(f"STARTING n = {n}")
    print("=" * 80)

    qpu_qubit_config = make_qpu_config(n, max_qubits_per_qpu=5)

    print(f"QPU config: {qpu_qubit_config}")
    print("Hyperparameters:")
    print(f"  depth = {depth}")
    print(f"  init_type = {init_type}")
    print(f"  lr = {lr}")
    print(f"  max_iters = {max_iters}")
    print(f"  rel_tol = {rel_tol}")
    print(f"  num_shots = {num_shots}")
    print(f"  final_shots = {final_shots}")
    print(f"  warm_start_population = {warm_start_population}")
    print(f"  warm_start_iters = {warm_start_iters}")
    print(f"  warm_start_shots = {warm_start_shots}")
    print(f"  energy_mode = {energy_mode}")
    print(f"  cvar_alpha = {cvar_alpha}")

    # ------------------------------------------------------------
    # Create one QUBO for this n
    # ------------------------------------------------------------

    qubo_seed = base_seed + 1000 * n

    qubo_start = time.time()

    Q, q_linear, z_exact, cost_exact, qubo_attempts = create_random_qubo_unique_solution(
        n=n,
        seed=qubo_seed,
        density=density
    )

    qubo_time = time.time() - qubo_start

    print(f"Unique QUBO found after {qubo_attempts} attempt(s)")
    print(f"Brute-force exact solution: {z_exact}")
    print(f"Brute-force exact cost: {cost_exact}")
    print(f"QUBO generation + brute-force time: {qubo_time:.4f} seconds")

    # ------------------------------------------------------------
    # Optional QUBO scaling for training stability
    # ------------------------------------------------------------

    scale = max(
        float(np.max(np.abs(Q))),
        float(np.max(np.abs(q_linear))),
        1.0
    )

    Q_train = Q / scale
    q_linear_train = q_linear / scale

    # ------------------------------------------------------------
    # Run distributed DVQE n3 times on same QUBO
    # ------------------------------------------------------------

    for run_id in range(1, n3 + 1):

        run_seed = base_seed + 1000 * n + run_id

        print(f"\nRunning DVQE for n={n}, run {run_id}/{n3}...")

        start = time.time()

        try:
            z_dvqe_raw, final_circuit, hist = dvqe(
                mode=mode,
                Q=Q_train,
                q_linear=q_linear_train,
                init_type=init_type,
                depth=depth,
                lr=lr,
                max_iters=max_iters,
                qpu_qubit_config=qpu_qubit_config,
                rel_tol=rel_tol,
                num_shots=num_shots,
                final_shots=final_shots,
                warm_start_population=warm_start_population,
                warm_start_iters=warm_start_iters,
                warm_start_shots=warm_start_shots,
                seed=run_seed,
                verbose=verbose,
                backend=fast_backend,
                energy_mode=energy_mode,
                cvar_alpha=cvar_alpha
            )

            elapsed = time.time() - start

            z_dvqe_raw = np.array(z_dvqe_raw, dtype=int)

            cost_dvqe_raw = float(
                z_dvqe_raw.T @ Q @ z_dvqe_raw + q_linear @ z_dvqe_raw
            )

            z_dvqe, cost_dvqe = local_bitflip_search(
                z_dvqe_raw,
                Q,
                q_linear,
                max_passes=20
            )

            gap_raw = cost_dvqe_raw - cost_exact
            gap = cost_dvqe - cost_exact

            exact_match_raw = bool(np.array_equal(z_dvqe_raw, z_exact))
            exact_match = bool(np.array_equal(z_dvqe, z_exact))

            status = "success"

        except Exception as e:
            elapsed = time.time() - start

            z_dvqe_raw = None
            z_dvqe = None
            cost_dvqe_raw = None
            cost_dvqe = None
            gap_raw = None
            gap = None
            exact_match_raw = False
            exact_match = False
            status = f"failed: {str(e)}"

        raw_results.append({
            "n": n,
            "run_id": run_id,
            "qpu_qubit_config": str(qpu_qubit_config),
            "backend": str(fast_backend),
            "depth": depth,
            "init_type": init_type,
            "lr": lr,
            "max_iters": max_iters,
            "rel_tol": rel_tol,
            "num_shots": num_shots,
            "final_shots": final_shots,
            "warm_start_population": warm_start_population,
            "warm_start_iters": warm_start_iters,
            "warm_start_shots": warm_start_shots,
            "energy_mode": energy_mode,
            "cvar_alpha": cvar_alpha,
            "qubo_seed": qubo_seed,
            "run_seed": run_seed,
            "qubo_scale": scale,
            "z_exact": z_exact.tolist(),
            "cost_exact": cost_exact,
            "z_dvqe_raw": None if z_dvqe_raw is None else z_dvqe_raw.tolist(),
            "cost_dvqe_raw": cost_dvqe_raw,
            "gap_raw": gap_raw,
            "exact_match_raw": exact_match_raw,
            "z_dvqe": None if z_dvqe is None else z_dvqe.tolist(),
            "cost_dvqe": cost_dvqe,
            "gap": gap,
            "exact_match": exact_match,
            "runtime_sec": elapsed,
            "status": status
        })

        print(f"Run {run_id} status: {status}")
        print(f"Runtime: {elapsed:.4f} seconds")
        print(f"Raw DVQE cost: {cost_dvqe_raw}")
        print(f"Raw gap: {gap_raw}")
        print(f"Raw exact match: {exact_match_raw}")
        print(f"Polished DVQE cost: {cost_dvqe}")
        print(f"Polished gap: {gap}")
        print(f"Polished exact match: {exact_match}")

    # ------------------------------------------------------------
    # Summary for this n
    # ------------------------------------------------------------

    successful_runs = [
        r for r in raw_results
        if r["n"] == n and r["status"] == "success"
    ]

    if len(successful_runs) > 0:
        runtimes_success = [r["runtime_sec"] for r in successful_runs]

        gaps_raw_success = [r["gap_raw"] for r in successful_runs]
        gaps_success = [r["gap"] for r in successful_runs]

        min_time = float(np.min(runtimes_success))
        max_time = float(np.max(runtimes_success))
        avg_time = float(np.mean(runtimes_success))
        median_time = float(np.median(runtimes_success))
        std_time = float(np.std(runtimes_success))

        min_gap_raw = float(np.min(gaps_raw_success))
        max_gap_raw = float(np.max(gaps_raw_success))
        avg_gap_raw = float(np.mean(gaps_raw_success))

        min_gap = float(np.min(gaps_success))
        max_gap = float(np.max(gaps_success))
        avg_gap = float(np.mean(gaps_success))
        median_gap = float(np.median(gaps_success))

        exact_match_percentage_raw = (
            100.0 * sum(r["exact_match_raw"] for r in successful_runs)
            / len(successful_runs)
        )

        exact_match_percentage = (
            100.0 * sum(r["exact_match"] for r in successful_runs)
            / len(successful_runs)
        )

    else:
        min_time = None
        max_time = None
        avg_time = None
        median_time = None
        std_time = None

        min_gap_raw = None
        max_gap_raw = None
        avg_gap_raw = None

        min_gap = None
        max_gap = None
        avg_gap = None
        median_gap = None

        exact_match_percentage_raw = 0.0
        exact_match_percentage = 0.0

    summary_results.append({
        "n": n,
        "n3_runs_requested": n3,
        "successful_runs": len(successful_runs),
        "qpu_qubit_config": str(qpu_qubit_config),
        "backend": str(fast_backend),
        "depth": depth,
        "init_type": init_type,
        "lr": lr,
        "max_iters": max_iters,
        "rel_tol": rel_tol,
        "num_shots": num_shots,
        "final_shots": final_shots,
        "warm_start_population": warm_start_population,
        "warm_start_iters": warm_start_iters,
        "warm_start_shots": warm_start_shots,
        "energy_mode": energy_mode,
        "cvar_alpha": cvar_alpha,
        "cost_exact": cost_exact,
        "z_exact": z_exact.tolist(),
        "min_runtime_sec": min_time,
        "max_runtime_sec": max_time,
        "avg_runtime_sec": avg_time,
        "median_runtime_sec": median_time,
        "std_runtime_sec": std_time,
        "min_gap_raw": min_gap_raw,
        "max_gap_raw": max_gap_raw,
        "avg_gap_raw": avg_gap_raw,
        "min_gap": min_gap,
        "max_gap": max_gap,
        "avg_gap": avg_gap,
        "median_gap": median_gap,
        "exact_match_percentage_raw": exact_match_percentage_raw,
        "exact_match_percentage": exact_match_percentage,
        "qubo_generation_bruteforce_time_sec": qubo_time,
        "qubo_attempts": qubo_attempts
    })

    print("\n" + "-" * 80)
    print(f"SUMMARY FOR n = {n}")
    print("-" * 80)
    print(f"Successful runs: {len(successful_runs)} / {n3}")
    print(f"Runtime min: {min_time}")
    print(f"Runtime max: {max_time}")
    print(f"Runtime avg: {avg_time}")
    print(f"Runtime median: {median_time}")
    print(f"Runtime std: {std_time}")
    print(f"Raw gap min: {min_gap_raw}")
    print(f"Raw gap max: {max_gap_raw}")
    print(f"Raw gap avg: {avg_gap_raw}")
    print(f"Polished gap min: {min_gap}")
    print(f"Polished gap max: {max_gap}")
    print(f"Polished gap avg: {avg_gap}")
    print(f"Polished gap median: {median_gap}")
    print(f"Raw exact match percentage: {exact_match_percentage_raw:.2f}%")
    print(f"Polished exact match percentage: {exact_match_percentage:.2f}%")
    print("-" * 80)


overall_time = time.time() - overall_start


# ============================================================
# FINAL TABLES
# ============================================================

raw_df = pd.DataFrame(raw_results)
summary_df = pd.DataFrame(summary_results)

print("\n\n" + "=" * 100)
print("FINAL SUMMARY TABLE")
print("=" * 100)

display(summary_df)

print("\n" + "=" * 100)
print("RAW RUN RESULTS")
print("=" * 100)

display(raw_df)

print(f"\nTotal runner time: {overall_time:.4f} seconds")


# ============================================================
# OPTIONAL CSV SAVE
# ============================================================

if save_csv:
    raw_df.to_csv(raw_results_csv, index=False)
    summary_df.to_csv(summary_csv, index=False)

    print("\nSaved CSV files:")
    print(raw_results_csv)
    print(summary_csv)
```
